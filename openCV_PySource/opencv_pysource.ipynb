```python
import cv2
image = cv2.imread("red_panda.jpg")
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow("Gray panda", gray_image)
cv2.imshow("Red panda", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Saving Image
cv2.imwrite("gray_panda.jpg", gray_image)
```

```python
#Read video from webcam
import cv2
import numpy as np
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
   
    cv2.imshow("frame", frame)
    
    key = cv2.waitKey(1)
    if key == 27:  #ESC Key
        break
cap.release()
cv2.destroyAllWindows()
```

```python
#Drawing & writing on Image
import cv2
import numpy as np
image = cv2.imread("red_panda.jpg")
shape = image.shape

# define the colors'
#The colors are in BGR format, it means that there are three channels, in order: blue, green and red.
blue = (255, 0, 0)
green = (0, 255, 0)
red = (0, 0, 255)
violet = (180, 0, 180)#B+R
yellow = (0, 180, 180)#G+R
white = (255, 255, 255)
```

```python
#We draw on the image in order: a line, a circle, a rectangle, an ellipse and a polygon.
cv2.line(image, (50, 30), (450, 35), blue, thickness=5)
cv2.line
cv2.circle(image, (240, 205), 23, red, -1)
cv2.rectangle(image, (50, 60), (450, 95), green, -1)#-1 -ve value for fillrectangle 
cv2.ellipse(image, (250, 150), (80, 20), 5, 0, 360, violet, -1)
points = np.array([[[140, 230], [380, 230], [320, 250], [250, 280]]], np.int32)
cv2.polylines(image, [points], True, yellow, thickness=3)
```

```python
#put some text on the image.
#We first define the font with which we want to display the text and then with the 
#function cv2.putText we display it on the screen.
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(image, "MYPanDa", (20, 180), font, 4, white)
```

```python
#Showing img on screen

cv2.imshow("red panda", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
